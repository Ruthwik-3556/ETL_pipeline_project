```python
# database.py
import sqlite3

def create_connection(db_name="company.db"):
    conn = sqlite3.connect(db_name)
    print(" Database connected")
    return conn

def create_table(conn):
    query = """
    CREATE TABLE IF NOT EXISTS employees (
        id INTEGER PRIMARY KEY,
        name TEXT,
        age INTEGER,
        department TEXT,
        salary REAL,
        joining_date TEXT
    )
    """
    conn.execute(query)
    conn.commit()
    print(" Table created")
```
